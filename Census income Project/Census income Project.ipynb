```python
df.columns
```

```python
for col in df.columns:
    if df[col].dtype!="object":
        plt.boxplot(df[col])
        plt.xlabel(col)
        plt.show()
```

```python
df["capital-gain"].value_counts()
```

```python
out_col=['hours-per-week','fnlwgt','age','education-num']
```

```python
# Outlier removal
for col in out_col:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    LW=Q1-1.5*IQR  #lower Wiskers
    UW=Q3+1.5*IQR #upper wiskers
    df=df[(df[col]>=LW) &(df[col]<=UW)]    
```

```python
df
```

# LAbel Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for col in df.columns:
    if df[col].dtype=="object":
        df[col]=le.fit_transform(df[col])  # labels will be assigned in alphabetical order
```

# MOdel Buiding 

```python
# Splitting the data into Indepencent and dependent
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df["income"]
```

```python
# split the data into training and testing 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=56)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo_model=LogisticRegression()
```

```python
# training the model
lo_model.fit(x_train,y_train)
```

```python
#testing
log_pred=lo_model.predict(x_test)
```

```python
log_pred
```

```python
# 0--> income<=50K
# 1--> income>50K
```

```python
y_test
```

```python
# Checking accuracy measures
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
cm=confusion_matrix(y_test,log_pred)
sns.heatmap(cm,annot=True,cmap="Blues",fmt="d")
```
